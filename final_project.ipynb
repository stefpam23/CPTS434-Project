### Pre-processing the data

```python
list_1=list(train.columns)
list_cate=[]
for i in list_1:
    if train[i].dtype=='object':
        list_cate.append(i)
le=LabelEncoder()
for i in list_cate:
    train[i]=le.fit_transform(train[i])
```

```python
X = train.drop('FILENAME',axis=1)
y = train['FILENAME']
```

## Training and Testing Data Based On Labels

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=12)
```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

## Tenserflow Processing

```python
model = tf.keras.Sequential([
      tf.keras.layers.Dense(units=80,activation='relu',name = 'input_layer'),
      tf.keras.layers.Dense(units=60,activation='relu',name = 'dense_layer1'),
      tf.keras.layers.Dense(units=40,activation='relu',name = 'dense_layer2'),
      tf.keras.layers.Dense(units=1,name='Output_layer')
],name='Model')   # Dropout and Batch Normalization can also be used.


        
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',        # In loss mae can also be used.
    metrics=['binary_accuracy'],
)


history = model.fit(X_train,y_train,batch_size=256,
    epochs=150,
)
```

Plotting Fitted line, training loss, and weight

```python
# Plotting training loss
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Get weights of the first dense layer
weights, biases = model.layers[1].get_weights()

# Plotting the weights
plt.figure(figsize=(10, 6))
plt.hist(weights.flatten(), bins=20)
plt.title('Weights Distribution of the First Dense Layer')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```

```python
# Make predictions
predictions = model.predict(X_train)

# Plotting fitted line
plt.figure(figsize=(10, 6))
plt.plot(y_train, predictions, 'ro')   # change 'ro' to change the color and type of the marker
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Fitted Line')
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], color='blue', linestyle='--')
# Limit the range of x and y axes
plt.xlim([0, 1000])
plt.ylim([-1500, 1000])
plt.show()
```
